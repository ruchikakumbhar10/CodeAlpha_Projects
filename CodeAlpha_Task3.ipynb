## DATA VISUALISATION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('scraped_quotes.csv')
df
```

```python
# Barplot – Top 10 Authors by Quote Count
plt.figure(figsize=(10, 5))
top_authors = df['Author'].value_counts().head(10)
sns.barplot(x=top_authors.values, y=top_authors.index, palette='coolwarm')
plt.title('Top 10 Most Quoted Authors')
plt.xlabel('Number of Quotes')
plt.ylabel('Author')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart – Top 5 Tags
from collections import Counter
```

```python
# Convert string tags to list and explode
tags_series = df['Tags'].str.split(', ').explode()
tags_series
```

```python
top_tags = dict(Counter(tags_series).most_common(5))
top_tags
```

```python
plt.figure(figsize=(6, 6))
plt.pie(top_tags.values(), labels=top_tags.keys(), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 5 Most Common Tags')
plt.show()
```

```python
# Wordcloud – Visualize Most Used Words in Quotes
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Example using your 'Quote' column
text = ' '.join(df['Quote'])
text
```

```python
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text)
wordcloud
```

```python
plt.figure(figsize=(5, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequent Words in Quotes')
plt.show()
```

```python
# Countplot – All Tags (Top 10)
plt.figure(figsize=(7, 3))
top_tag_series = tags_series.value_counts().head(10)
sns.barplot(x=top_tag_series.values, y=top_tag_series.index, palette='flare')
plt.title("Top 10 Tags")
plt.xlabel("Count")
plt.ylabel("Tag")
plt.tight_layout()
plt.show()
```
